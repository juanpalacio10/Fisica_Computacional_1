```python
#Cargo y limpio los datos
#df = pd.read_csv("/content/millas_por_galon.csv")
df = pd.read_csv("/home/anferivera/Dropbox/UDEA/Fisica_Computacional_1/Sesiones/seaborn/millas_por_galon.csv")

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
df = df.dropna()
```

```python
df[["mpg","horsepower"]].describe()
#Esta me muestra media desviacion estandar,minimo, maximo, cuartiles
```

```python
corr = df["mpg"].corr(df["horsepower"])
print("Coeficiente de correlación:", corr)
#Basandome en la correlacion de pearson que es entre -1<=r<=1
#Que en este caso es de aprox -0.75 lo que indica una correlacion negativa fuerte
#osea que a mayor potencia menor eficiencia de combustible
```

```python
#Para la regresión calculo la recta mpg=  a+b*horsepower
coef = np.polyfit(df["horsepower"], df["mpg"], 1)

pendiente = coef[0]
intercepto = coef[1]

print("Pendiente:", pendiente)
print("Intercepto:", intercepto)
```

```python
y_pred = pendiente*df["horsepower"] + intercepto

ss_res = np.sum((df["mpg"] - y_pred)**2)
ss_tot = np.sum((df["mpg"] - df["mpg"].mean())**2)

r2 = 1 - ss_res/ss_tot

print("R^2 =", r2)
#Este r2 indica que porcentaje de la variabilidad de mpg explica el horsepower
```

```python
plt.figure(figsize=(8,6))

sns.regplot(
    data=df,
    x="horsepower",
    y="mpg",
    ci=95
)

plt.title("Relación mpg vs horsepower con ajuste lineal (95% C.L.)")
plt.xlabel("Potencia (horsepower)")
plt.ylabel("Millas por galón (mpg)")

plt.show()
```
